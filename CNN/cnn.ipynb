```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dataset_path="C:/Users/Bunny/OneDrive/Documents/techworks/CNN/PetImages"
```

```python
# Image preprocessing 
datagen = ImageDataGenerator(

    #normalize pixel 
    rescale=1./255, 
    
    # Data augmentation
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,

    #split data
    validation_split=0.2) 
```

```python
# load training data
# Train data 

train_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
```

```python
# load validation data

val_data= datagen.flow_from_directory(
    dataset_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
print(train_data.class_indices)
```

```python
# Building the CNN model
model=keras.Sequential()
```

```python
# First convolutional layer

model.add(
    keras.layers.Conv2D(
        filters=32, 
        kernel_size=(3, 3), 
        activation='relu', 
        input_shape=(128, 128, 3)
    )
)
```

```python
# First pooling layer

model.add(
    keras.layers.MaxPooling2D(pool_size=(2, 2))
)
```

```python
# 2nd convolutional layer
model.add(
    keras.layers.Conv2D(
        filters=64, 
        kernel_size=(3, 3), 
        activation='relu'
    )
)
```

```python
# 2nd pooling layer
model.add(
    keras.layers.MaxPooling2D(pool_size=(2, 2))
)
```

```python
# Flattening layer

model.add(keras.layers.Flatten())
```

```python
# Dense layer

model.add(
    keras.layers.Dense(
        units=128, activation='relu'
        ))
```

```python
# output layer

model.add(
    keras.layers.Dense(
        units=1, activation='sigmoid'
    )
)
```

```python
import tensorflow as tf
print(tf.config.list_physical_devices())


# Compile CNN
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)


```

```python
# Train the model
history=model.fit(
    train_data,
    epochs=2,
    validation_data=val_data
)
```

```python
loss, accuracy = model.evaluate(val_data)
print(f"Validation Loss: {loss:.4f}")
print(f"Validation Accuracy: {accuracy:.4f}")
```
